```python
#change directory new folder
os.chdir('D:\Data Machinelearning\Transjakarta\CSVTJFinal')
```

```python
#list all files csv in directory
csv_files = [f for f in os.listdir() if f.endswith('.csv')]
```

```python
#read all csv files
dfsf = []

for csv in csv_files:
    df = pd.read_csv(csv)
    dfsf.append(df)
```

```python
#concat all csv files
final_df= pd.concat(dfsf, ignore_index=True)
```

```python
#checking drop zero value
clean_df = final_df.drop(final_df[final_df['jumlah_penumpang']==0].index)
```

```python
#sorting data to bulan
clean_df.sort_values(by=['bulan'], inplace=True,ignore_index=True)
```

```python
#checking info data
clean_df.info()
```

```python
#checking info data
clean_df.tail()
```

```python
#table to csv
clean_df.to_csv('DataTransjakarta.csv', index=False)
```

```python
#spliting variable trayek with delimiter '-'
splited_df = clean_df['trayek'].str.split(' - ', expand=True)
```

```python
#checking the value
splited_df
```

```python
#spliting variable trayek with delimiter 'via'
splited_df2=splited_df[1].str.split(' via ', expand=True)
```
